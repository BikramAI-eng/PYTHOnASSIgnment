```python
#Project 1
#Create a CLI (Command Line Interface) contact book that allows users to:
#● Add a new contact (append to file)
#● View all contacts (read from file)
#● Search for a contact (read and filter)
#● Handle file-related exceptions (e.g., file not found)
#File Used:
#contacts.txt (stores contact info: Name, Phone)Create a simple banking system that:


FILE_NAME = "contacts.txt"

def add_contact():
    """Add a new contact to the file"""
    name = input("Enter name: ")
    phone = input("Enter phone: ")
    try:
        with open(FILE_NAME, "a") as f:
            f.write(f"{name},{phone}\n")
        print(" Contact added successfully!")
    except Exception as e:
        print(" Error writing to file:", e)

def view_contacts():
    """Display all contacts stored in the file"""
    try:
        with open(FILE_NAME, "r") as f:
            contacts = f.readlines()
            if not contacts:
                print("No contacts found.")
            else:
                print("\n--- Contact List ---")
                for contact in contacts:
                    name, phone = contact.strip().split(",")
                    print(f"Name: {name}, Phone: {phone}")
    except FileNotFoundError:
        print(" File not found. Please add a contact first!")

def search_contact():
    """Search a contact by name or phone"""
    keyword = input("Enter name or phone to search: ")
    try:
        with open(FILE_NAME, "r") as f:
            found = False
            for line in f:
                if keyword in line:
                    name, phone = line.strip().split(",")
                    print(f"Found → Name: {name}, Phone: {phone}")
                    found = True
            if not found:
                print("No match found.")
    except FileNotFoundError:
        print(" File not found. Please add a contact first!")

# Main Menu Loop
while True:
    print("\n--- Contact Book ---")
    print("1. Add Contact")
    print("2. View All Contacts")
    print("3. Search Contact")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_contact()
    elif choice == "2":
        view_contacts()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        print("Exiting Contact Book. Goodbye!")
        break
    else:
        print("Invalid choice. Try again.")
```

```python
#Project 2
#● Stores customer info in a file
#● Allows deposits and withdrawals using functions
#● Updates customer balance
#● Logs all transactions in a separate file
#● Handles exceptions gracefully
#Files Used:
#customers.txt — stores customer records in the format:
#Name,AccountNumber,Balance
#transactions.txt — appends every deposit or withdrawal record with timestamp
# Contact Book Program


from datetime import datetime

CUSTOMER_FILE = "customers.txt"
TRANSACTION_FILE = "transactions.txt"

def load_customers():
    """Load all customers into a dictionary from the file"""
    customers = {}
    try:
        with open(CUSTOMER_FILE, "r") as f:
            for line in f:
                name, acc_no, balance = line.strip().split(",")
                customers[acc_no] = {"name": name, "balance": float(balance)}
    except FileNotFoundError:
        open(CUSTOMER_FILE, "w").close()  # create empty file
    return customers

def save_customers(customers):
    """Save updated customer details back into the file"""
    with open(CUSTOMER_FILE, "w") as f:
        for acc_no, data in customers.items():
            f.write(f"{data['name']},{acc_no},{data['balance']}\n")

def log_transaction(acc_no, action, amount):
    """Log each transaction with date and time"""
    with open(TRANSACTION_FILE, "a") as f:
        f.write(f"{datetime.now()} - {acc_no} - {action} - {amount}\n")

def create_account(customers):
    """Create a new customer account"""
    name = input("Enter customer name: ")
    acc_no = input("Enter account number: ")
    balance = float(input("Enter initial balance: "))
    customers[acc_no] = {"name": name, "balance": balance}
    save_customers(customers)
    log_transaction(acc_no, "Account Created", balance)
    print(" Account created successfully!")

def deposit(customers):
    """Deposit money into an account"""
    acc_no = input("Enter account number: ")
    if acc_no in customers:
        amount = float(input("Enter deposit amount: "))
        customers[acc_no]["balance"] += amount
        save_customers(customers)
        log_transaction(acc_no, "Deposit", amount)
        print(" Deposit successful!")
    else:
        print(" Account not found.")

def withdraw(customers):
    """Withdraw money from an account"""
    acc_no = input("Enter account number: ")
    if acc_no in customers:
        amount = float(input("Enter withdrawal amount: "))
        if customers[acc_no]["balance"] >= amount:
            customers[acc_no]["balance"] -= amount
            save_customers(customers)
            log_transaction(acc_no, "Withdrawal", amount)
            print(" Withdrawal successful!")
        else:
            print(" Insufficient balance.")
    else:
        print(" Account not found.")

# Main Menu
customers = load_customers()

while True:
    print("\n--- Banking System ---")
    print("1. Create Account")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        create_account(customers)
    elif choice == "2":
        deposit(customers)
    elif choice == "3":
        withdraw(customers)
    elif choice == "4":
        print("Exiting Banking System. Goodbye!")
        break
    else:
        print("Invalid choice. Try again.")
```
